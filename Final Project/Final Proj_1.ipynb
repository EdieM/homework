```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("2015_02_clickstream.tsv", sep='\t', header=0)
#not using ID's, so discarded
df = df[['prev_title', 'curr_title', 'n', 'type']]
df.columns = ['prev', 'curr', 'n', 'type']
df.head(20)
```

```python
#here looking at top pages/issues accessed
df.groupby('curr').sum().sort('n', ascending=False)[:25]
```

```python
#here looking at top referers
df.groupby('prev').sum().sort('n', ascending=False)[:30]
```

```python
#what was trending on fb
df_facebook = df[df['prev'] == 'other-facebook']
df_facebook.groupby('curr').sum().sort('n', ascending=False)[:25]
```

```python
#what was trending on the twits
df_twitter = df[df['prev'] == 'other-twitter']
df_twitter.groupby('curr').sum().sort('n', ascending=False)[:25]
```

```python
#what was trending, but didn't exist
#either the pages had been deleted, or they haven't been written yet, but someone is holding the link
df_redlinks = df[df['type'] == 'redlink']
df_redlinks.groupby('curr').sum().sort('n', ascending=False)[:25]
```
