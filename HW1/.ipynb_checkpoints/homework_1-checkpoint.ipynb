```python
# 1. How many passengers are in our passenger list? 
# From here forward, we’ll assume our dataset represents the full passenger list for the Titanic.

passenger_num = len(table_data.index)
print "There were " + str(passenger_num) + " total passengers on the Titanic."
```

```python


# 2. What is the overall survival rate?
# I'm assuming 1 = survived, and 0 = did not survive

survived_num = len(table_data[table_data['Survived'] == 1])
total_survival = (float("{0:.2f}".format(((survived_num) / float(passenger_num)) * 100)))
print "There were " + str(survived_num) + " total survivors on the Titanic. That is a survival rate of " + str(total_survival) + "%."
```

```python


# 3. How many male passengers were onboard?

male_num = len(table_data[table_data['Sex'] == 'male'])
print "There were " + str(male_num) + " male passengers on the Titanic." 
```

```python

# 4. How many female passengers were onboard?

female_num = len(table_data[table_data['Sex'] == 'female'])
print "There were " + str(female_num) + " female passengers on the Titanic." 
```

```python

# 5. What is the overall survival rate of male passengers?

male_survivors = len(table_data[ (table_data['Survived'] == 1) & (table_data['Sex'] == 'male') ])
total_male_survival = (float("{0:.2f}".format(((male_survivors) / float(male_num)) * 100)))
print str(male_survivors) + " men survived the crash. That is a " + str(total_male_survival) + "% survival rate."
```

```python

# 6. What is the overall survival rate of female passengers?

female_survivors = len(table_data[ (table_data['Survived'] == 1) & (table_data['Sex'] == 'female') ])
total_female_survival = (float("{0:.2f}".format(((female_survivors) / float(female_num)) * 100)))
print str(female_survivors) + " women survived the crash. That is a " + str(total_female_survival) + "% survival rate."
```
